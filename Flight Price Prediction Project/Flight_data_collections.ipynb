#### * visting the mentioned url

```python
url = "https://flight.yatra.com/air-search-ui/dom2/trigger?type=O&viewName=normal&flexi=0&noOfSegments=1&origin=DEL&originCountry=IN&destination=BOM&destinationCountry=IN&flight_depart_date=19%2F01%2F2023&ADT=1&CHD=0&INF=0&class=Economy&source=fresco-flights&unqvaldesktop=1318631119934"
driver.get(url)
time.sleep(5)
```

```python
Airline_name = []
Depart_time = []
Arival_time = []
Duration = []
Stoppage = []
Price = []
From = []
To = []

#scraping the Airline_name
try:
    air_name = driver.find_elements(By.XPATH,'//span[@class="i-b text ellipsis"]')
    for os in air_name:
        Airline_name.append(os.text)
except NoSuchElementException:
    Airline_name.append('-')

    
#scraping the data of Departure time
try:
    air_dec = driver.find_elements(By.XPATH,'//div[@class="i-b pr"]')
    for os in air_dec:
        Depart_time.append(os.text)
except NoSuchElementException:
    Depart_time.append('-')


#scraping data of the Arival time
try:
    air_ari = driver.find_elements(By.XPATH,'//p[@class="bold fs-15 mb-2 pr time"]')
    for disp in air_ari:
        Arival_time.append(disp.text)
except NoSuchElementException:
    Arival_time.append("-")


# scraping data of Duration of the flight
try:
    dur = driver.find_elements(By.XPATH,'//p[@class="fs-12 bold du mb-2"]')
    for pro in dur:
        Duration.append(pro.text)
except NoSuchElementException:
    Duration.append('-')
    
    
# scraping the number of stoppage
try:
    stop = driver.find_elements(By.XPATH,'//div[@class=" font-lightgrey fs-10 tipsy i-b fs-10"]')
    for pro in stop:
        Stoppage.append(pro.text)
except NoSuchElementException:
    Stoppage.append('-')
    
    
#scraping the data of fare of
try:
    op_ser = driver.find_elements(By.XPATH,'//div[@class="i-b tipsy fare-summary-tooltip fs-18"]')
    for os in op_ser:
        Price.append(os.text)
except NoSuchElementException:
    Price.append('-')
    
    
# scraping  boarding location 
try:
    name = driver.find_elements(By.XPATH,"//p[1][@class='fs-10 font-lightgrey no-wrap city ellipsis']")
    for i in name:
        From.append(i.text)
except NoSuchElementException:
    From.append('-')
    
    
# scraping  landing location
try:
    name = driver.find_elements(By.XPATH,"//p[2][@class='fs-10 font-lightgrey no-wrap city ellipsis']")
    for i in name:
        To.append(i.text)
except NoSuchElementException:
    To.append('-')   
```

```python
import pandas as pd
df1 = pd.DataFrame()
df1['Airline name']=Airline_name
df1['From']=From
df1['To']=To
df1['Arival time']=Arival_time
df1['Depart time']=Depart_time
df1['Stoppage']=Stoppage
df1['Duration']=Duration
df1['Price']=Price
df1
```

```python
df1.to_csv('df1')
```
